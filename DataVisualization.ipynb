```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/sample_data/mtcars.csv')
df
```

```python
# barplot using column cyl  and hp

sns.barplot(x=df['cyl'], y = df['hp'], palette='Set3')
plt.xlabel('Cylinder')
plt.ylabel('Horse Power')
plt.title('Barplot using column cyl and hp')
plt.show()
```
