```python
for col in df.columns:
  if df[col].dtype=='object':
    df[col]=df[col].fillna(df[col].mode()[0])
  else:
    df[col]=df[col].fillna(df[col].mean())
```

```python
df.columns
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
# df.drop_duplicates(inplace= True)
```

```python
for col in df.columns:
  if(df[col].dtype!= 'object'):
    plt.boxplot(df[col])
    plt.xlabel(col)
    plt.show()
```

```python
out_list=['bmi','past_consultations','Hospital_expenditure','Anual_Salary']
for col in out_list:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)

  IQR=Q3-Q1

  LB=Q1-1.5*(IQR)
  UB=Q3+1.5*(IQR)

  df=df[(df[col]>=LB) & (df[col]<=UB)]
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
for col in df.columns:
  if(df[col].dtype == 'object'):
     df[col] = LE.fit_transform(df[col])
```

```python
df.info()
```

```python
# model building
# 1.splitting the data x,y
# 2.spliting in terms of testing and training
# 3.model initilization
# 4.train the model
# 5.predictions
# 6.evaluate
```

```python
x=df.iloc[:,:-1]
y=df['charges']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
results=pd.DataFrame(columns=['Actual_values','Predicted_values'])
results['Actual_values']=y_test
results['Predicted_values']=y_pred
results
```

```python
from sklearn.metrics import *
```

```python
# r square
result=r2_score(y_test,y_pred)
result*100
```

```python
sns.regplot(x=y_pred,y=y_test)
plt.title('Regression Plot')
plt.xlabel('Model Answers')
plt.ylabel('Actual Values')
plt.show()
```
