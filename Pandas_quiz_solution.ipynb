```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('/content/sample_data/housing.csv')
df
```

```python
# dataframe column info
df.info()
```

```python
# convert datatime column to date right format
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# checking the null in data set
df.isnull().sum()
```

```python
# highest bed room
filtered_df = df.loc[df['bedrooms'] == df['bedrooms'].max(), :]
filtered_df
```

```python
# highest price
highest_price = df.loc[df['price'] == df['price'].max(), :]
highest_price
```

```python
# hughest 5 prices by date
df.groupby('price')['bedrooms'].max()
```

```python
# duplicated rows
df.duplicated().sum()
```

```python
df.sort_values('date', ascending=False)
```
