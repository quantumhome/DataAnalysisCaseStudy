2) Creating two hypothesis ( null hypothesis , Alternate hypothesis )

3) Applying the appropriate test ( based on requirement or case )

if **p_value is > 0.05** = Fail to reject the null hypothesis ( Accept null hypothesis )

if **p_value is < 0.05** = Reject the null hypothesis ( Accept the alternate hypothesis )


# a) It was observed that the average purchase made by the Men of the age 18-25 was 10000. Is it still the same?

**A one-sample t-test** is a statistical test used to determine whether the mean of a single sample is significantly different from a known or hypothesized population mean.

It helps in testing hypotheses about the average value of a dataset.

**Types of t-tests:**

**One-sample t-test**: Compares the sample mean to a known value (population mean).

**Independent two-sample t-test**: Compares the means of two independent groups.

**Paired sample t-test**: Compares means from the same group at different times (e.g., before and after treatment).

Key Concepts:

Null Hypothesis (H₀): The sample mean is equal to the population mean.

Alternative Hypothesis (H₁): The sample mean is different (greater or less) than the population mean.

```python
# We are going to use one-sample t-test
```

```python
# Extract the data of age group 18-25 ( because the age 0-17 was encoded as 0 , so the age 18-25 is encoded as 1 )

new_data = data[(data["Age"]==1) & (data["Gender"]==1)]
new_data
```

```python
# Collect sample from population
sample =new_data.sample( 3600 , random_state=5)
sample
```

```python
new_data["Purchase"].mean() # Average purchase of population
```

```python
# Null hypothesis = mean will be equal to 10,000
# Alternate hypothesis = mean will not be equal to 10,000
```

```python
from scipy.stats import ttest_1samp
a_mean=10000  # Population measn
```

```python
t_stat , p_value = ttest_1samp(sample["Purchase"] ,a_mean)
print(t_stat, p_value)
```

```python
value = 1.7977812377777705e-10
print(f"{value:.25f}")  # Adjust the number of decimal places as needed
```

**The p-value** tells you the probability of observing the sample data, or something more extreme, assuming the null hypothesis is true.

It helps in making a decision about the null hypothesis.

**t-statistic (t_stat):**

The t-statistic is a measure of how much the sample mean deviates from the population mean, expressed in terms of the standard error. It helps in determining whether the difference between the sample mean and the hypothesized population mean is significant.

**A larger t-statistic** (positive or negative) indicates a greater deviation from the population mean.

**A small t-statistic** indicates that the sample mean is close to the population mean

```python
#if p_value is > 0.05 = Fail to reject the null hypothesis ( Accept null hypothesis )
#if p_value is < 0.05 = Reject the null hypothesis ( Accept the alternate hypothesis )
```

As our p_value is less than 0.05 os **we are rejecting the null hypothesis** and **accepting the alternate hypothesis**

**Average purchase is not equal to 10,000**

# b) It was observed that the percentage of women that spend more than 10000 was 35%. Is it still the same?

You will use a **one-sample z-tes**t for proportions to compare the observed proportion of women spending more than 10,000 with the hypothesized proportion of 35%.

**Why Use a One-Sample Z-Test for Proportions?**

A z-test for proportions is used when you want to compare an observed sample proportion to a known or hypothesized population proportion.

```python
#null hypothesis - proportion is 35%.
#alternate hypothesis - proportion is not 35%.

data_new = data[(data['Purchase'] > 10000)]
```

```python
data_new
```

```python
data_new["Gender"].value_counts()
```

```python
#No of women in the sample
count=data_new['Gender'].value_counts()[0]

#number of observations
nobs = len(data_new['Gender'])

# hypothesised value
p0 = 0.35
```

```python
count,nobs,p0
```

```python
18685/91057
```

```python
#one sample z test
from statsmodels.stats.proportion import proportions_ztest
```

```python
z_stat, p_val = proportions_ztest(count=count,nobs=nobs,value=p0)
print(z_stat, p_val)
```

**z-statistic**: Measures how many standard errors the observed proportion is away from the hypothesized proportion of 35%.

**p-value**: If the p-value is less than 0.05, reject the null hypothesis and conclude that the proportion of women spending more than 10,000 is different from 35%.

```python
#p-value is less than 0.05, reject the null hypothesis.
#the proportion of women spending more than 10000, is not 35%.
```

# d) Is the average purchase made by men and women of the age 18-25 same

**we can perform an independent two-sample t-test**

```python
#null hypothesis - spends are equal
#alternate hypothesis - spends are not equal

data_men = data.loc[(data['Gender'] == 1)& (data['Age'] == 1)]
data_women = data.loc[(data['Gender'] == 0) & (data['Age'] == 1)]
```

```python
#creating samples
data_men_sample = data_men.sample(500, random_state=0)
data_women_sample = data_women.sample(500, random_state=0)
```

```python
#checking variances of the two samples
print(data_men_sample.Purchase.var())
print(data_women_sample.Purchase.var())
```

```python
#sample means
print(data_men_sample.Purchase.mean())
print(data_women_sample.Purchase.mean())
```

```python
#computing f statistic
from scipy.stats import f
F = data_men_sample.Purchase.mean()/data_women_sample.Purchase.mean()
print(F)
```

```python
#calculating the degrees of freedom
df1 = len(data_men_sample) - 1
df2 = len(data_women_sample) - 1

print(df1, df2)

# degrees of freedom in the F-test indicate the number of independent pieces of information available for estimating variances.
```

```python
#p-value
import scipy

p_val=1-scipy.stats.f.cdf(F, df1, df2)
p_val

# Varivances of the two samples are not equal
```

```python
#two sample t-test for unequal variances
from scipy.stats import ttest_ind

t_stat_2, p_val_2 = ttest_ind(data_men_sample["Purchase"], data_women_sample["Purchase"], equal_var=False)
print(t_stat_2, p_val_2)
```
